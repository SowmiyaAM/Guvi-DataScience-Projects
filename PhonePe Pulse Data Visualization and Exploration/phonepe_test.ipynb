```python
import os
import json 
import pandas as pd
import mysql.connector
```

AGGREGATED DATA 

```python
# agg_trans - State list

path1 = "E:/AMS Docs/Courses/Guvi/Data Science/Projects/02_PhonePe Project/PhonePe_Code/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_state_list = os.listdir(path1)

columns1 = {"States":[], "Years":[], "Quarters":[], "Transaction Type":[], "Transaction Count":[], "Transaction Amount":[]}

# agg_trans - Extracting the required details from all documents and stored in dataframe

for state in agg_trans_state_list:
    cur_state = path1+state+"/"
    agg_trans_year_list = os.listdir(cur_state)
    for year in agg_trans_year_list:
        cur_year=cur_state+year+"/"
        agg_trans_file_list = os.listdir(cur_year)
        for file in agg_trans_file_list:
            cur_file = cur_year+file
            data=open(cur_file,"r")

            AT=json.load(data)   #converts nested json string into python dictionary
            for i in AT["data"]["transactionData"]:
                columns1["Transaction Type"].append(i["name"])
                columns1["Transaction Count"].append(i["paymentInstruments"][0]["count"])
                columns1["Transaction Amount"].append(i["paymentInstruments"][0]["amount"])
                columns1["States"].append(state)
                columns1["Years"].append(year)
                columns1["Quarters"].append(int(file.strip(".json")))

agg_trans = pd.DataFrame(columns1)
agg_trans
```

```python
# agg_users - State list

path2 = "E:/AMS Docs/Courses/Guvi/Data Science/Projects/02_PhonePe Project/PhonePe_Code/pulse/data/aggregated/user/country/india/state/"
agg_users_state_list = os.listdir(path2)

columns2 = {"States":[], "Years":[], "Quarters":[], "Brands":[], "Users":[], "Percentage":[]}

# agg_users - Extracting the required details from all documents and stored in dataframe

for state in agg_users_state_list:
    cur_state=path2+state+"/"
    agg_users_year_list = os.listdir(cur_state)
    for year in agg_users_year_list:
        cur_year=cur_state+year+"/"
        agg_users_file_list = os.listdir(cur_year)
        for file in agg_users_file_list:
            cur_file=cur_year+file
            data=open(cur_file,"r")

            AU=json.load(data)
            try:
                for i in AU["data"]["usersByDevice"]:
                    columns2["Brands"].append(i["brand"])
                    columns2["Users"].append(i["count"])
                    columns2["Percentage"].append(i["percentage"])
                    columns2["States"].append(state)
                    columns2["Years"].append(year)
                    columns2["Quarters"].append(int(file.strip(".json")))
            except:
                pass

agg_users = pd.DataFrame(columns2)
agg_users           
```

DATA CLEANING - Manipulating "States" Column in all the 9 dataframes

```python
docs = [agg_trans,agg_users,agg_ins,map_trans,map_users,map_ins,top_trans,top_users,top_ins]
for i in docs:
    i["States"] = i["States"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")
    i["States"] = i["States"].str.replace("-"," ")
    i["States"] = i["States"].str.title()
    i["States"] = i["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu","Dadra and Nagar Haveli and Daman and Diu")
```
